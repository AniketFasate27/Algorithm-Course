```python
import numpy as np
import matplotlib.pyplot as plt

minReceivedPower_dBm = -60  # Minimum received power in dBm
pathLossExponent = 3.2  # Path loss exponent for urban areas
frequency_Hz = 900e6  # Frequency in Hz
speedOfLight = 3e8  # Speed of light in m/s
wavelength = speedOfLight / frequency_Hz  # Wavelength
refPathLoss = 20 * np.log10(4 * np.pi * 1 / wavelength)  # Path loss at 1 meter
minReceivedPower_Watts = 10 ** ((minReceivedPower_dBm - 30) / 10)
distances = np.arange(1, 101)  # in meters
requiredTransmitPower_dBm = np.zeros(len(distances))

for i in range(len(distances)):
    pathLoss = 10 * pathLossExponent * np.log10(distances[i]) + refPathLoss
    requiredTransmitPower_Watts = minReceivedPower_Watts * 10 ** (pathLoss / 10)
    requiredTransmitPower_dBm[i] = 10 * np.log10(requiredTransmitPower_Watts) + 30

plt.plot(distances, requiredTransmitPower_dBm)
plt.xlabel('Distance (meters)')
plt.ylabel('Transmission Power (dBm)')
plt.title('Required Transmission Power vs Distance')
plt.grid(True)
plt.show()
```
