```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
d0 = 1  
pl_d0 = 40  # path loss @ 1m
path_loss_exp = 2
min_rx_pwr_dbm = 0 # from previous calculation

# Calculate Tx Power over Distance
dist = np.linspace(1, 3000, 100)  
path_loss = pl_d0 + 10*path_loss_exp*np.log10(dist/d0)
req_tx_pwr = min_rx_pwr_dbm + path_loss

# Plot 
plt.plot(dist, req_tx_pwr, '-b', label='Tx Power')
plt.xlabel('Distance (m)')
plt.ylabel('Required Tx Power (dBm)')
plt.title('Tx Power vs Distance')
plt.legend()
plt.grid()
plt.show()
```
