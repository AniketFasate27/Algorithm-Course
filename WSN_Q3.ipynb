```python
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate transmitted power based on Friis equation
def calculate_transmitted_power(distance):
    received_power = -80  # Received power in dBm
    antenna_gain_tx = 0  # Transmitting antenna gain in dB
    antenna_gain_rx = 0  # Receiving antenna gain in dB
    system_loss = 0  # System loss in dB

    transmitted_power = received_power + 20 * np.log10(distance) - antenna_gain_tx - antenna_gain_rx + system_loss
    # print(distance)
    # print(transmitted_power) 
    return transmitted_power

# Generate distances from 1 to 500 meters
distances = np.arange(1, 501, 1)
# distances = 50
# Calculate transmitted power for each distance

transmitted_powers = [calculate_transmitted_power(d) for d in distances]

 
# Plotting
plt.plot(distances, transmitted_powers)
plt.title('Transmitted Power vs Distance')
plt.xlabel('Distance (meters)')
plt.ylabel('Transmitted Power (dBm)')
plt.grid(True)
plt.show()
```
