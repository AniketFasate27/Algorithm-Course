```python
import numpy as np 
from matplotlib import pyplot as plt

# Given Parameters
bandwidth = 1e6 # 1 MHz
data_rate = 300e3 # 300 kbps
noise_dbm = -80
req_snr_db = 20

# Calculations 
noise_pwr = np.power(10, noise_dbm/10) # in W
noise_bw_dbm = noise_dbm + 10*np.log10(bandwidth)
noise_bw = np.power(10, noise_bw_dbm/10) 

min_rx_pwr_dbm = noise_bw_dbm + req_snr_db
min_rx_pwr = np.power(10, min_rx_pwr_dbm/10)

# Print result
print(f"Min RX Power: {min_rx_pwr_dbm:.2f} dBm, {min_rx_pwr:.3} W")

# Visualization
x = [0, 1e-4]
y = [noise_pwr, min_rx_pwr]
plt.plot(x,y)
plt.xlabel("Time (s)")
plt.ylabel("Power (W)")
plt.title("Min RX Power Requirement")
plt.xticks([0, 1e-4], [0, "1 Msg Tx Time"]) 
plt.grid(True)
plt.show()
```
