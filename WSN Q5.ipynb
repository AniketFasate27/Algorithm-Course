```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
d0 = 1 
pl_d0 = 40 # path loss at reference distance 
n = 2 # path loss exponent
pr_req = -80 # required received power 

# Calculate transmit power over distances
d = np.linspace(10, 5000, 100)
pl = pl_d0 + 10*n*np.log10(d/d0)
pt = pr_req + pl

# Plot 
fig, ax = plt.subplots()
ax.plot(d, pt)

ax.set(xlabel='Distance (m)', 
       ylabel='Transmit Power (dBm)',
       title='Transmit Power vs Distance')
ax.grid()

fig.tight_layout()
plt.show()
```
