```python
import matplotlib.pyplot as plt

# Parameters
max_tx_pwr = 10 # dBm
ap_distance = 150 # meters 
max_range = 100 # meters (calculated previously)

# Distances from AP
distances = range(0,200,10)

# Calculate no. of hops 
hops = [int(round(d/max_range)) for d in distances]

# Plot
fig, ax = plt.subplots()
ax.plot(distances, hops)
ax.set_xlabel("Distance from AP (m)")  
ax.set_ylabel("Number of hops")
ax.set_title("Hops vs Distance from AP")
ax.grid(True)

# Annotate points
for x,y in zip(distances, hops):
    label = str(y)
    ax.annotate(label, # this is the text
                (x,y), # this is the point to label
                textcoords="offset points", # how to position the text
                xytext=(0,10), # distance from text to points (x,y)
                ha='center') # horizontal alignment can be left, right or center

plt.show()
```
